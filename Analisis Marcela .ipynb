# Panorama laboral general 

```python
# pie chart, femenino vs masculino per estado
def autopct_format(values):
        def my_format(pct):
            total = sum(values)
            val = int(round(pct*total/100.0))
            return '{:.1f}%\n({v:d})'.format(pct, v=val)
        return my_format
    
states=df['ESTADO'].unique()
regimen=df['REGIMEN'].unique()

for state in states: 
    #sexo 
    df_estado=df[df['ESTADO']==state]
    
    s=df_estado['sexo'].value_counts() #distribucion de sexo por estado
    plt.pie(s,labels = s.index, autopct=autopct_format(s))
    plt.title(state)
    plt.figure()
    #regimen
    r=df_estado['REGIMEN'].value_counts() #distribucion de regimen por estado
    plt.pie(r,labels = r.index, autopct=autopct_format(s))
    plt.figure()
    #regimen y sexo 
    df_counts=df_estado[['sexo', 'REGIMEN']].value_counts().reset_index(name='counts')
    pivot = pd.pivot_table(data=df_counts, index=['REGIMEN'], columns=['sexo'], values='counts')
    ax = pivot.plot.bar(stacked=True, color =['r', 'b'], figsize=(8,6))
    plt.figure()
    
    # histograma de edad por sexo en cada estado 
    sns.kdeplot(df_estado.loc[(df_estado['sexo']=='FEMENINO'),
            'EDAD'], color='r', shade=True, label='FEMENINO')
    sns.kdeplot(df_estado.loc[(df_estado['sexo']=='MASCULINO'),
            'EDAD'], color='b', shade=True, label='MASCULINO')
    plt.legend()
    
    #kernel density estimation: de edad por regimen en cada estado 
    plt.figure()
    colors=['r', 'b','g','y']
    i=0
    for reg in regimen: 
        sns.kdeplot(df_estado.loc[(df_estado['REGIMEN']==reg),
            'EDAD'], color=colors[i], shade=True, label=reg)
        i+=1 
    plt.legend()
    
    
    
    
    plt.figure()
    
    
```

```python
def leer_datos_por_año(archivo, columna_estado, columnas):
    datos_por_año = {}
    xls = pd.ExcelFile(archivo)
    
    datos_por_año=pd.DataFrame(columns=columnas)
    for hoja in xls.sheet_names:
        df = pd.read_excel(archivo, sheet_name=hoja)
        año = int(hoja)
        df['AÑO'] = año  # Agregar columna de año para identificar de qué año es cada registro
        
        # Crear la columna 'PEN_ALIM' con valores predeterminados
        df['PEN_ALIM'] = 'SI'
        
        datos_por_año=pd.concat([datos_por_año,df])
        #datos_por_año[año] = df  # pd stack 
    return datos_por_año


columnas= ['EPS','sexo', 'ESTADO', 'REGIMEN', 'JORNA', 'DEPARTAMENTO','EDAD', 'FECHAA','AÑO', 'PEN_ALIM']
df_pension_VO= leer_datos_por_año('Pensión alimenticia 2014.2022 Veracruz Oaxaca PTRI.xlsx', 'ESTADO', columnas)
```

```python
# prepare to merge, modify columns, etc. 

columnas_cambio={'GENERO':'sexo', 'TIPO CONTRATACION': 'REGIMEN', 'JORNADA':'JORNA'}

df_VO_completo=df_VO_completo.rename(columns=columnas_cambio)

df_VO_completo
```

```python
#cambiar departamento y agregar colunmna con descripcion 
#regex to extract middle number from df_VO_completo['DEPARTAMENTO']

df_VO_completo['DEPT DESC']=df_VO_completo['DEPARTAMENTO'].str.extract('\s\S*\s(.*)')

df_VO_completo['DEPARTAMENTO']=df_VO_completo['DEPARTAMENTO'].str.extract('^.*?\s(\w+)')
```

```python
#change sexo and make it uppercase 
df_VO_completo['sexo'] = df_VO_completo['sexo'].str.upper()
#change regimen and rename with Dictionary 
plazas= {'Planta Sindicalizado':'PS' , 'Transitorio Sindicalizado':'TS',
       'Planta Confianza':'PC', 'Transitorio Confianza': 'TC'}

df_VO_completo['REGIMEN']=df_VO_completo['REGIMEN'].replace(plazas, regex=True)

```
